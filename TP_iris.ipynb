```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)
iris = load_iris()
```

```python
names = iris['target_names']
names
```

```python
dtIris = pd.DataFrame(iris['data'])
dtIris
```

```python
target = iris['target']
target
```

```python
dtIris['cible'] = target
dtIris
```

```python
dtIris.cible.value_counts()
```

```python
dtIris.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
dtIrisScaled = scaler.fit_transform(dtIris.iloc[:,0:4])
dtIrisScaled
FinalIris = pd.DataFrame(dtIrisScaled)
FinalIris
```

```python
irisN = FinalIris
irisN['Species']= target
irisN.columns = ['Sepal_len','Sepal_wid','Petal_len','Petal_wid','Species']
irisN
```

```python
from sklearn import neighbors
```

```python
KNN=neighbors.KNeighborsClassifier(n_neighbors=5,
                             weights='uniform',
                             algorithm='kd_tree',
                             leaf_size=30,
                             p=2,
                             metric='minkowski',
                             metric_params=None,
                             n_jobs=None)
```

```python
X=irisN[["Sepal_len","Sepal_wid","Petal_len","Petal_wid"]]
y=irisN['Species']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
y
```

```python
KNN.fit(X_train,y_train)
```

```python
KNN.score(X_train,y_train)
```
